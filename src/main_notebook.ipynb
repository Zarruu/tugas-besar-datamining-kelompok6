```python
# Import fungsi dari modul yang telah dibuat di dalam folder src/
from src.data_loader import load_raw_data, save_processed_data
from src.utils import preprocess_data, find_elbow
from src.model import evaluate_k_ranges, fit_final_kmeans
import matplotlib.pyplot as plt
```

```python
# 1. Load data mentah
df = load_raw_data("../data/raw/dataset_pulau_jawa.csv")
df.head()
```

```python
# 2. Preprocessing & Feature Scaling
df_clean, X_normalized = preprocess_data(df)
print(f"Shape data setelah normalisasi: {X_normalized.shape}")
```

```python
# 3. Cari K Optimal & Evaluasi
k_range, inertia, silhouette = evaluate_k_ranges(X_normalized)
best_k = find_elbow(k_range, inertia)
print(f"K Optimal terdeteksi secara matematis: K = {best_k}")

# Visualisasi Metode Elbow
plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o', linestyle='--')
plt.axvline(x=best_k, color='red', linestyle=':', label=f'K optimal = {best_k}')
plt.title('Metode Elbow untuk K-Means')
plt.xlabel('Jumlah Cluster (K)')
plt.ylabel('Inertia (WCSS)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 4. Fitting Model Final (Misalnya diputuskan menggunakan K=3)
df_final, model = fit_final_kmeans(df_clean, X_normalized, n_clusters=3)
df_final[['title', 'city', 'Cluster_Name_Final']].head(10)
```

```python
# 5. Menyimpan Hasil Akhir clustering ke folder processed
save_processed_data(df_final, "../data/processed/hasil_clustering_pulau_jawa_lengkap.csv")
```
