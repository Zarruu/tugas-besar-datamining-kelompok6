```python
import pandas as pd

print("1. Memuat dataset...")
df = pd.read_csv('raw_dataset_pulau_jawa.csv')
print(f"Total baris awal: {len(df)}")

# --- A. CLEANING & STANDARISASI TEKS ---
print("\n2. Standarisasi singkatan kota/kabupaten...")
city_mapping = {
    'Kota Bks': 'Kota Bekasi',
    'Jkt Utara': 'Jakarta Utara', 'Jkt Selatan': 'Jakarta Selatan',
    'Jkt Barat': 'Jakarta Barat', 'Jkt Timur': 'Jakarta Timur', 'Jkt Pusat': 'Jakarta Pusat'
}
df['city'] = df['city'].replace(city_mapping)

print("3. Translasi format Inggris ke Indonesia...")
df['city'] = df['city'].str.replace(r'^(.*?)\sCity$', r'Kota \1', regex=True)
df['city'] = df['city'].str.replace(r'^(.*?)\sRegency$', r'Kabupaten \1', regex=True)

# --- B. FILTERING DOMAIN & KOLOM ---
print("4. Filtering kota di luar Pulau Jawa...")
non_jawa_cities = ['KOTA'] # Sesuaikan jika ada nama kota luar Jawa yang spesifik
df = df[~df['city'].isin(non_jawa_cities)].copy()
print(f"   Baris setelah filter: {len(df)}")

print("\n5. Drop kolom tidak diperlukan...")
kolom_dibuang = ['imageUrl', 'categories', 'url', 'website', 'phone', 'street', 'state', 'countryCode']
df.drop(columns=kolom_dibuang, errors='ignore', inplace=True)

# --- C. SELEKSI FITUR & HANDLING MISSING VALUES ---
print("\n6. Seleksi fitur untuk K-Means...")
kolom_penting = ['title', 'categoryName', 'city', 'totalScore', 'reviewsCount']
df = df[kolom_penting].copy()

print("7. Menghapus baris dengan nilai null...")
df.dropna(subset=['totalScore', 'reviewsCount', 'city', 'categoryName'], inplace=True)
print(f"   Baris setelah hapus null: {len(df)}")

# --- D. FILTERING BISNIS / THRESHOLD ---
print("8. Filter reviewsCount >= 100...")
df = df[df['reviewsCount'] >= 100.0].copy()
df.reset_index(drop=True, inplace=True)
print(f"   Baris final: {len(df)}")

# --- E. VERIFIKASI & SIMPAN ---
print("\n9. Verifikasi akhir:")
print(df.isnull().sum())

nama_file = 'dataset_pulau_jawa_clean.csv'
df.to_csv(nama_file, index=False)
print(f"\n✅ Preprocessing selesai! Disimpan sebagai '{nama_file}'")
```
