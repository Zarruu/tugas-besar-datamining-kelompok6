```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data dulu
df = pd.read_csv('raw_dataset_pulau_jawa.csv')

print("=== EDA AWAL ===")
# 1. Info struktur data & tipe kolom
print("\n📊 Info Dataset:")
print(df.info())

# 2. Statistik deskriptif (kolom numerik)
print("\n📈 Statistik Deskriptif:")
print(df[['totalScore', 'reviewsCount']].describe())

# 3. Cek missing values awal
print("\n❌ Missing Values Awal:")
print(df.isnull().sum())

# 4. Distribusi kategori & kota
print("\n🏙️ 10 Kota Terbanyak:")
print(df['city'].value_counts().head(10))

print("\n📂 10 Kategori Terbanyak:")
print(df['categoryName'].value_counts().head(10))

# 5. Visualisasi distribusi skor & jumlah review
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.histplot(df['totalScore'].dropna(), bins=20, kde=True)
plt.title('Distribusi totalScore')

plt.subplot(1, 2, 2)
sns.histplot(df['reviewsCount'].dropna(), bins=30, kde=True)
plt.title('Distribusi reviewsCount')
plt.tight_layout()
plt.show()

# 6. Scatter plot hubungan score vs review (opsional)
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='reviewsCount', y='totalScore', alpha=0.5)
plt.title('totalScore vs reviewsCount')
plt.xscale('log') # log scale karena review biasanya skew
plt.show()
```
